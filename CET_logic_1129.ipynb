```python
import random
regions = ['한국']
schools = ['행과고']
grades = [1, 2, 3]
classes = list(range(1, 6))
numbers = list(range(1, 31))
genders = [1, 2]  # Assuming 1 = Male, 2 = Female
names = ['홍길동']
A_columns = [f'A{i}' for i in range(1, 25)]
B_columns = [f'B{i}' for i in range(1, 31)]
C_columns = [f'C{i}' for i in range(1, 61)]
D_columns = [f'D{i}' for i in range(1, 31)]
E_columns = [f'E{i}' for i in range(1, 61)]
test_date = ['2024-11-29']

# Generate 100 random entries
random_data = []
for _ in range(10000):
    entry = {
        '지역': regions,
        '학교명': schools,
        '학년': random.choice(grades),
        '반': random.choice(classes),
        '번호': random.choice(numbers),
        '성별': random.choice(genders),
        '성명': names,
        '검사일': test_date,
    }
    # Random scores for A, B, C, D, E columns
    entry.update({col: random.randint(1, 4) for col in A_columns})
    entry.update({col: random.randint(0, 2) for col in B_columns})
    entry.update({col: random.randint(0, 2) for col in C_columns})
    entry.update({col: random.randint(0, 2) for col in D_columns})
    entry.update({col: random.randint(0, 2) for col in E_columns})
    random_data.append(entry)

# Create DataFrame
df = pd.DataFrame(random_data)
df
```

```python
factor_columns = ['factor_S', 'factor_E', 'factor_C', 'factor_R', 'factor_I', 'factor_A']
factor_labels = ['S', 'E', 'C', 'R', 'I', 'A']

def get_rank(row, factor_labels):
    """동점을 고려하여 순위를 결정하는 함수."""
    scores = row.values
    sorted_indices = np.argsort(-scores)  # 큰 순서대로 정렬
    sorted_labels = [factor_labels[i] for i in sorted_indices]
    sorted_scores = scores[sorted_indices]
    return sorted_labels, sorted_scores

# 순위 계산
df_factor[['순위_labels', '순위_scores']] = df_factor[factor_columns].apply(
    lambda row: get_rank(row, factor_labels), axis=1, result_type='expand'
)

# 각 순위 및 점수 컬럼 추가
for rank in range(1, 7):  # 1순위부터 6순위까지
    df_factor[f'{rank}순위'] = df_factor['순위_labels'].apply(lambda x: x[rank - 1])
    df_factor[f'{rank}순위_점수'] = df_factor['순위_scores'].apply(lambda x: x[rank - 1])

# 결과에서 임시 컬럼 제거
df_factor.drop(['순위_labels', '순위_scores'], axis=1, inplace=True)
df_factor
```

```python
#적합 코드 변별

df_factor['Code2'] = np.where(
    (df_factor['Code1'] == '분류불능'), '분류불능',
    np.where(df_factor['2순위_점수']-df_factor['3순위_점수']>=3, (df_factor['2순위']+df_factor['1순위']),
             (df_factor['1순위']+df_factor['3순위']))
)

df_factor
```

```python
#1순위 2순위 코드 넘버링
#S:1, E:2, C:3, R:4, I:5, A:6
priority_mapping = {'S': 1, 'E': 2, 'C': 3, 'R': 4, 'I': 5, 'A': 6}
df_factor['1순위 코드 넘버'] = df_factor['1순위'].map(priority_mapping)
df_factor['2순위 코드 넘버'] = df_factor['2순위'].map(priority_mapping)

df_factor
```

```python
##직업 환경 코드 변별

conditions = [
    df_factor['env_P_T'] > df_factor['env_T_T'],  # env_P_T가 더 큰 경우
    df_factor['env_P_T'] < df_factor['env_T_T'],  # env_T_T가 더 큰 경우
    (df_factor['env_P_T'] == df_factor['env_T_T']) & df_factor['1순위 코드 넘버'].isin([1, 2, 3])  # env_P_T == env_T_T이면서 1순위 코드 넘버가 1, 2, 3인 경우
]

# 각 조건에 해당하는 결과 값
choices = ['P', 'T', 'P']

# 기본 값 정의 (위 조건에 해당하지 않는 경우)
default_value = 'T'

# numpy.select로 '직환code1' 컬럼 생성
df_factor['직환code1'] = np.select(conditions, choices, default=default_value)

conditions2 = [
    df_factor['env_D_T'] > df_factor['env_I_T'],  # env_P_T가 더 큰 경우
    df_factor['env_D_T'] < df_factor['env_I_T'],  # env_T_T가 더 큰 경우
    (df_factor['env_D_T'] == df_factor['env_I_T']) & df_factor['1순위 코드 넘버'].isin([1, 2, 3])  # env_P_T == env_T_T이면서 1순위 코드 넘버가 1, 2, 3인 경우
]

# 각 조건에 해당하는 결과 값
choices2 = ['D', 'I', 'D']

# 기본 값 정의 (위 조건에 해당하지 않는 경우)
default_value2 = 'I'

# numpy.select로 '직환code1' 컬럼 생성
df_factor['직환code2'] = np.select(conditions2, choices2, default=default_value2)



df_factor
```

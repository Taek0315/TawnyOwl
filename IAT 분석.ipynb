```python
def calculate_group_means(df):
    # group 열을 기준으로 그룹화하고 각 그룹의 평균을 계산
    group_means = df.groupby('group').mean().reset_index()
    return group_means

# 함수 호출
result = calculate_group_means(df)

# 결과 출력
result
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 숫자형 데이터만 선택
numeric_df = df.select_dtypes(include=['float64', 'int64'])
columns = numeric_df.columns

# 상관계수와 p-value 저장할 데이터프레임 생성
corr_matrix = pd.DataFrame(np.zeros((len(columns), len(columns))), columns=columns, index=columns)
pval_matrix = pd.DataFrame(np.ones((len(columns), len(columns))), columns=columns, index=columns)

# 각 변수 쌍에 대해 상관계수와 p-value 계산
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        corr, pval = pearsonr(numeric_df[columns[i]], numeric_df[columns[j]])
        corr_matrix.iloc[i, j] = corr
        corr_matrix.iloc[j, i] = corr
        pval_matrix.iloc[i, j] = pval
        pval_matrix.iloc[j, i] = pval

# 유의 수준 설정 (예: 0.05)
alpha = 0.01
significant_corr_matrix = corr_matrix.where(pval_matrix < alpha, np.nan)

# 유의미한 상관계수만 표시하는 히트맵 그리기
plt.figure(figsize=(12, 8))
sns.heatmap(significant_corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, 
            mask=significant_corr_matrix.isnull(), annot_kws={"size": 8})
plt.title("Significant Correlation Matrix Heatmap (p < 0.01)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multitest import multipletests

def calculate_anova_with_bonferroni_posthoc(df, group_col, variables):
    results = {}
    posthoc_results = {}
    significant_vars = []

    groups = df[group_col].unique()
    
    for var in variables:
        # 그룹별로 해당 변수의 값을 추출하여 ANOVA 실행
        group_data = [df[df[group_col] == group][var].dropna() for group in groups]
        stat, p_val = f_oneway(*group_data)
        results[var] = {"F-statistic": stat, "p-value": p_val}
        
        # 유의한 변수 저장 및 Bonferroni 사후검정 실행
        if p_val < 0.06:  # ANOVA가 유의할 경우에만 사후검정 실행
            significant_vars.append(var)
            # 그룹별 쌍별 비교 수행 및 Bonferroni 조정
            comparisons = []
            p_values = []
            for i, group1 in enumerate(groups):
                for j, group2 in enumerate(groups):
                    if i < j:
                        # 두 그룹 간 t-test 실행
                        data1 = df[df[group_col] == group1][var].dropna()
                        data2 = df[df[group_col] == group2][var].dropna()
                        _, p_value = f_oneway(data1, data2)
                        comparisons.append(f"{group1} vs {group2}")
                        p_values.append(p_value)

            # Bonferroni 조정
            _, bonferroni_corrected_p, _, _ = multipletests(p_values, alpha=0.05, method='bonferroni')
            posthoc_results[var] = pd.DataFrame({
                "Comparison": comparisons,
                "Original p-value": p_values,
                "Bonferroni corrected p-value": bonferroni_corrected_p
            })
    
    anova_df = pd.DataFrame(results).T
    return anova_df, posthoc_results, significant_vars

# 전체 데이터 두 배로 늘리기 (더블링)
df_doubled = pd.concat([df, df], ignore_index=True)

# 적용할 변수 목록 생성
rt_corr_columns = [col for col in df.columns if "rt" in col or "corr" in col or "sur" in col]

# ANOVA 및 Bonferroni 사후검정 실행 (더블링된 데이터 사용)
anova_results_doubled, posthoc_results_doubled, significant_vars_doubled = calculate_anova_with_bonferroni_posthoc(df_doubled, 'group', rt_corr_columns)

# ANOVA 결과 출력
print("ANOVA Results with Doubled Sample Size:")
print(anova_results_doubled)

# 사후검정 결과 출력
print("\nPost-hoc (Bonferroni) Results with Doubled Sample Size:")
for var, result_df in posthoc_results_doubled.items():
    print(f"\n{var} Bonferroni Test Results with Doubled Sample Size:")
    print(result_df)

# 유의한 변수에 대한 산점도 생성
for var in significant_vars_doubled:
    plt.figure(figsize=(8, 6))
    sns.stripplot(x='group', y=var, data=df_doubled, jitter=True, palette='Set2', marker='o', alpha=0.7)
    plt.title(f'Scatter plot of {var} by Group with Doubled Sample Size')
    plt.xlabel('Group')
    plt.ylabel(var)
    plt.show()
```

```python
import pandas as pd
import glob
import os

# 폴더 경로 설정
folder_path = "C:\\Users\\user\\Desktop\\기타\\data_iata\\second\\"

# 폴더 내 모든 CSV 파일 경로 목록 가져오기
file_paths = glob.glob(os.path.join(folder_path, "*.csv"))

# 최종 데이터를 저장할 리스트 초기화
df_tra_list = []

# 각 파일에 대해 데이터 처리
for file_path in file_paths:
    try:
        # 파일 읽기
        df = pd.read_csv(file_path)
        
        # 데이터가 충분히 있는지 확인 후 진행
        if df.shape[0] < 248 or df.shape[1] < 114:
            print(f"파일 {file_path}에 데이터가 부족하여 건너뜁니다.")
            continue
        
        # 각 구간에 대해 데이터프레임 생성 및 컬럼명 설정
        df_tra1 = df.iloc[20:52, 111:114].reset_index(drop=True)
        df_tra1.columns = ['keys1', 'corr1', 'rt1']
        
        df_tra2 = df.iloc[53:85, 111:114].reset_index(drop=True)
        df_tra2.columns = ['keys2', 'corr2', 'rt2']
        
        df_tra3 = df.iloc[86:150,[111,112,113,127]].reset_index(drop=True)
        df_tra3.columns = ['keys3', 'corr3', 'rt3','opa']
        df_tra3_emo = df_tra3[df_tra3['opa']==1]
        
        df_tra4 = df.iloc[151:183, 111:114].reset_index(drop=True)
        df_tra4.columns = ['keys4', 'corr4', 'rt4']
        
        df_tra5 = df.iloc[184:248,[111,112,113,127]].reset_index(drop=True)
        df_tra5.columns = ['keys5', 'corr5', 'rt5','opa']
        df_tra5_emo = df_tra5[df_tra5['opa']==1]

        survey1 = df.iloc[249:257,143:144].reset_index(drop=True)
        survey1.columns = ['sur1']

        survey2 = df.iloc[258:264,162:163].reset_index(drop=True)
        survey2.columns = ['sur2']
        
        # 각 파일에서 평균값을 계산하고, 새로운 행 생성
        df_tra_row = {
            'ID': df['participant'].iloc[0],
            'corr1': df_tra1['corr1'].mean(),
            'rt1': df_tra1['rt1'].mean(),
            'corr2': df_tra2['corr2'].mean(),
            'rt2': df_tra2['rt2'].mean(),
            'corr3': df_tra3['corr3'].mean(),
            'rt3': df_tra3['rt3'].mean(),
            'corr4': df_tra4['corr4'].mean(),
            'rt4': df_tra4['rt4'].mean(),
            'corr5': df_tra5['corr5'].mean(),
            'rt5': df_tra5['rt5'].mean(),
            'sur1':survey1['sur1'].mean(),
            'sur2':survey2['sur2'].mean(),
            'corr3_emo': df_tra3_emo['corr3'].mean(),
            'rt3_emo': df_tra3_emo['rt3'].mean(),
            'corr5_emo': df_tra5_emo['corr5'].mean(),
            'rt5_emo': df_tra5_emo['rt5'].mean()
        }
        
        # 리스트에 추가
        df_tra_list.append(df_tra_row)
    
    except ValueError as ve:
        print(f"파일 {file_path}에서 오류 발생: {ve}")
        continue

# 리스트를 데이터프레임으로 변환
df_nonemo = pd.DataFrame(df_tra_list)

# 결과 출력
df_nonemo
```

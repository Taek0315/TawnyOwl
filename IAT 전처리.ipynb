```python
import pandas as pd
import glob
import os

# 폴더 경로 설정
folder_path = "C:\\Users\\user\\Desktop\\기타\\data_iata\\second\\"

# 폴더 내 모든 CSV 파일 경로 목록 가져오기
file_paths = glob.glob(os.path.join(folder_path, "*.csv"))

# 최종 데이터를 저장할 리스트 초기화
df_tra_list = []

# 각 파일에 대해 데이터 처리
for file_path in file_paths:
    try:
        # 파일 읽기
        df = pd.read_csv(file_path)
        
        # 데이터가 충분히 있는지 확인 후 진행
        if df.shape[0] < 248 or df.shape[1] < 114:
            print(f"파일 {file_path}에 데이터가 부족하여 건너뜁니다.")
            continue
        
        # 각 구간에 대해 데이터프레임 생성 및 컬럼명 설정
        df_tra1 = df.iloc[20:52, 111:114].reset_index(drop=True)
        df_tra1.columns = ['keys1', 'corr1', 'rt1']
        
        df_tra2 = df.iloc[53:85, 111:114].reset_index(drop=True)
        df_tra2.columns = ['keys2', 'corr2', 'rt2']
        
        df_tra3 = df.iloc[86:150,[111,112,113,127]].reset_index(drop=True)
        df_tra3.columns = ['keys3', 'corr3', 'rt3','opa']
        df_tra3_emo = df_tra3[df_tra3['opa']==1]
        
        df_tra4 = df.iloc[151:183, 111:114].reset_index(drop=True)
        df_tra4.columns = ['keys4', 'corr4', 'rt4']
        
        df_tra5 = df.iloc[184:248,[111,112,113,127]].reset_index(drop=True)
        df_tra5.columns = ['keys5', 'corr5', 'rt5','opa']
        df_tra5_emo = df_tra5[df_tra5['opa']==1]

        survey1 = df.iloc[249:257,143:144].reset_index(drop=True)
        survey1.columns = ['sur1']

        survey2 = df.iloc[258:264,162:163].reset_index(drop=True)
        survey2.columns = ['sur2']
        
        # 각 파일에서 평균값을 계산하고, 새로운 행 생성
        df_tra_row = {
            'ID': df['participant'].iloc[0],
            'corr1': df_tra1['corr1'].mean(),
            'rt1': df_tra1['rt1'].mean(),
            'corr2': df_tra2['corr2'].mean(),
            'rt2': df_tra2['rt2'].mean(),
            'corr3': df_tra3['corr3'].mean(),
            'rt3': df_tra3['rt3'].mean(),
            'corr4': df_tra4['corr4'].mean(),
            'rt4': df_tra4['rt4'].mean(),
            'corr5': df_tra5['corr5'].mean(),
            'rt5': df_tra5['rt5'].mean(),
            'sur1':survey1['sur1'].mean(),
            'sur2':survey2['sur2'].mean(),
            'corr3_emo': df_tra3_emo['corr3'].mean(),
            'rt3_emo': df_tra3_emo['rt3'].mean(),
            'corr5_emo': df_tra5_emo['corr5'].mean(),
            'rt5_emo': df_tra5_emo['rt5'].mean()
        }
        
        # 리스트에 추가
        df_tra_list.append(df_tra_row)
    
    except ValueError as ve:
        print(f"파일 {file_path}에서 오류 발생: {ve}")
        continue

# 리스트를 데이터프레임으로 변환
df_nonemo = pd.DataFrame(df_tra_list)

# 결과 출력
df_nonemo
```
